```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
def simularExp(a, mu):
    N = [100, 200, 400, 600, 800]
    sup = np.zeros(len(N))
    inf = np.zeros(len(N))
    norm = sps.norm(loc =0, scale=1)
    for i in range(len(N)):
        t = sps.t(df=N[i]-1)
        sample = np.random.exponential(mu, N[i])
        sigma = np.std(sample, ddof=1)
        Xn = np.mean(sample)
        sup[i] = t.ppf(1-a/2) * sigma / np.sqrt(N[i]) + Xn
        inf[i] = t.ppf(a/2) * sigma / np.sqrt(N[i]) + Xn
    return sup, inf, N
```

```python
def ej1(a):
    mu = 1/2
    sup, inf, N = simularExp(a, mu)
    for i in range(len(N)):
        plt.plot((N[i], N[i]), (sup[i], inf[i]), 'g-' if (mu >= inf[i] and mu <= sup[i]) else 'r-')
    plt.plot(np.linspace(0, 1000, 1000), np.ones(1000)*mu, 'b-')
    plt.show()

ej1(0.05)
```

```python
def ej3():
    lambd = 1/2
    bw = 0.15
    N = [100, 200, 300, 400]
    for i in range(len(N)):
        string = "N = " + str(N[i])
        sample = np.random.exponential(lambd, N[i])
        plt.hist(sample, bins= int((max(sample)-min(sample))/bw),density=True)
        plt.title(string)
        plt.show()

ej3()
```

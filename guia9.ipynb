```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ej1(muestra):
    esperanza = np.mean(muestra)
    varianza = np.var(muestra)
    esperanza2 = 0
    for i in muestra:
        esperanza2 += i
    esperanza2 = esperanza2 / len(muestra)
    varianza2 = 0
    for i in muestra:
        varianza2 += (i - esperanza2)**2
    varianza2 = varianza2 / (len(muestra)-1)
    
    return esperanza, varianza, esperanza2, varianza2
```

```python
def ej2(n):
    muestra = np.random.exponential(1/0.5,n)
    esp, var, esp2, var2 = ej1(muestra)
    print ("Esperanza2: ", esp2)
    print ("Varianza2: ", var2)
    return muestra

ej2(10)
ej2(30)
ej2(100)
```

```python
def ej3(muestra, bw):            # Ancho de banda
    sample = muestra# Nos quedamos con primeras 200 realicaciones 
    sample = np.sort(muestra)    # Ordeno la muestra

    plt.xlabel(r'$Y$', fontsize = 15)
    plt.ylabel(r'Frecuencia relativa', fontsize = 15)
    plt.hist(sample, bins=np.arange(min(sample), max(sample) + bw, bw), weights=np.zeros(len(sample)) + 1. / len(sample))  
    # Doy el peso de cada elemento para asegurarme que el histograma sea de frecuencia relativa (atenci√≥n: no se hace por default)

    plt.show()
    
muestra = ej2(100)
ej3(muestra, 0.4)
```
